```python
# Dependencies
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()

# URL of page to be scraped
url = 'https://stats.nba.com/players/bio/?Season=2018-19&SeasonType=Regular%20Season'

driver.get(url)

#capture the next arrow elements by reference their class 
elements = driver.find_elements_by_class_name("stats-table-pagination__next")

#obtain the page html using the .html attribute of the browser object
html = driver.page_source

# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(html, 'html.parser')

data_df = pd.read_html(str(soup.table))[0]

for x in range(1, 15):
    try:
#         elements = driver.find_elements_by_class_name("stats-table-pagination__next")
        elements[1].click()
    
        html = driver.page_source

        # Create BeautifulSoup object; parse with 'html.parser'
        soup = BeautifulSoup(html, 'html.parser')
        
        add_data_df = pd.read_html(str(soup.table))[0]

        data_df = data_df.append(add_data_df, ignore_index=True)
    except:
        pass
```

```python
heights = list(data_df['Height'])
for i in range(len(heights)):
    height_split = heights[i].split('-')
    
    feet = int(height_split[0])
    inches = int(height_split[1])
    
    total_height = (feet * 12 ) + inches
    
    data_df['Height'][i] = total_height
```

# NHL Player Bio Data (Fox Website) Scrape

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import time
import pandas as pd


# Use the read_html function in Pandas to automatically scrape any tabular data from a page.
data_url = "https://www.foxsports.com/nhl/players?teamId=0&season=2018&position=0&page=1&country=0&grouping=0&weightclass=0"

#using the pandas .read_html() method, we can convert the html tables found the page to pandas dataframes.
# Note the result of the .read_html() method is a list of dataframes pulled from the page
url_tables = pd.read_html(data_url, encoding= "utf-8")

# Reference the first element of our list of dataframes
NHL_Bio_df = url_tables[0]


executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)


# URL of page to be scraped
url = 'https://www.foxsports.com/nhl/players?teamId=0&season=2018&position=0&page=2&country=0&grouping=0&weightclass=0'

# using the .visit() method tell the browser to visit the url
browser.visit(url)
for x in range(1, 69):
    try:
        url_tables = pd.read_html(browser.url, encoding= "utf-8")
        
        # Reference the first element of our list of dataframes
        Bio_df = url_tables[0]
        
        NHL_Bio_df = NHL_Bio_df.append(Bio_df, ignore_index=True)
        
        browser.click_link_by_partial_text("Next")
    except:
        pass
```

```python
NHL_Bio_df.to_csv('../DataFrames/NHL_Bio.csv')
```

# NHL time on Ice Scraping Code(2018-2019)

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import time
import pandas as pd


# Use the read_html function in Pandas to automatically scrape any tabular data from a page.
data_url = "http://www.espn.com/nhl/statistics/player/_/stat/timeonice/sort/avgTimeOnIce"

#using the pandas .read_html() method, we can convert the html tables found the page to pandas dataframes.
# Note the result of the .read_html() method is a list of dataframes pulled from the page
url_tables = pd.read_html(data_url, encoding= "utf-8")

# Reference the first element of our list of dataframes
NHL_Time_df = url_tables[0]


executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)


# URL of page to be scraped
url = 'http://www.espn.com/nhl/statistics/player/_/stat/timeonice/sort/avgTimeOnIce/count/41'


player_count = 41
# using the .visit() method tell the browser to visit the url
browser.visit(url)
for x in range(1, 25):
    try:
        url_tables = pd.read_html(browser.url, encoding= "utf-8")
        
        # Reference the first element of our list of dataframes
        Time_df = url_tables[0]
        
        NHL_Time_df = NHL_Time_df.append(Time_df, ignore_index=True)
        player_count += 40
        next_page_url = "http://www.espn.com/nhl/statistics/player/_/stat/timeonice/sort/avgTimeOnIce/count/{}".format(player_count)
        browser.visit(next_page_url)
    except:
        pass
```

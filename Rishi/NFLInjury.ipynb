```python
import pandas as pd
import urllib.request

from bs4 import BeautifulSoup
```

```python
# scrape players injiured stats by cummulative season 2018

source = urllib.request.urlopen('https://www.spotrac.com/nfl/injured-reserve/2018/').read()
soup = BeautifulSoup(source, "html.parser")
table = soup.find_all('table', attrs={'class':'datatable'})
```

```python
c=0
for tb in table:
#     print(tb)
   c=c+1
print(c)
```

```python
first_table=table[4]
table_rows = first_table.find_all('tr')

res = []
for tr in table_rows:
   td = tr.find_all('td')
   row = [tr.text.strip() for tr in td if tr.text.strip()]
   if row:
       res.append(row)
injNFLdf_18 = pd.DataFrame(res)
```

```python
# Use the read_html function in Pandas to automatically scrape any tabular data from a page.
data_url = "https://www.foxsports.com/nfl/players"

#using the pandas .read_html() method, we can convert the html tables found the page to pandas dataframes.
# Note the result of the .read_html() method is a list of dataframes pulled from the page
url_tables = pd.read_html(data_url, encoding= "utf-8")

# Reference the first element of our list of dataframes
player_df = url_tables[0]
player_df
```

```python
player_df["Player"]= player_df['Player'].str.split(', ').str[1]
player_df
```

```python
player_df.to_csv()
```

```python
import pandas as pd
import urllib.request

from bs4 import BeautifulSoup
```

```python
# scrape players injured by cummulative season 2018

source = urllib.request.urlopen('http://www.espn.com/nhl/statistics/player/_/stat/timeonice/sort/avgTimeOnIce/year/2018').read()
soup = BeautifulSoup(source, "html.parser")
table = soup.find_all('table', attrs={'class':'datatable'})
```

```python
c=0
for tb in table:
#     print(tb)
    c=c+1
print(c)
```

```python
first_table=table[4]
table_rows = first_table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
df_18 = pd.DataFrame(res)
# print(df.count)

#print(df)
```

```python
df_18 = pd.DataFrame(res, columns=["PLAYER", "POS", "TEAM", "REASONS", "WEEKS", "2018 CAP", "2018 CASH"])

df_18.head()
```

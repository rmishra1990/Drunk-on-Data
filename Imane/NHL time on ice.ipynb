```python
import pandas as pd
```

```python
file1 = "NHL_Time_on_Ice_2019.csv"


NHLTime = pd.read_csv(file1)
NHLTime.head()
```

```python
NHLTime[['PLAYER','POS']] = NHLTime['PLAYER'].str.split(',',expand=True)

NHLTime.head()
```

```python
# flip first last name to last first
NHLTime["PLAYER"] = [" ".join(n.split(" ")[::-1]) for n in NHLTime["PLAYER"]]

NHLTime.head()
```

```python
# drop duplicated rows if any
NHLTime = NHLTime.drop_duplicates(subset=None, keep='first', inplace=False)
# print(injNHLdata18.shape)
NHLTime.head()
```

```python
NHLTime.shape
```

```python
# drop rows with na if any
NHLTime.dropna(how='any', inplace=True)
```

```python
NHLTime.shape
```

```python
NHLTime.to_csv('NHLTime.csv', index=True, header=True)
```

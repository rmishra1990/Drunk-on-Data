```python
import pandas as pd
import numpy as np
```

```python
file1 = "NHL_Injury_Database_data.csv"
injNHLdata = pd.read_csv(file1)
```

```python
injNHLdata.head()
```

```python
# DROP DUPLICATE COLUMN (GAMES MISSED) FILTER FOR 2018/19 INJ YRS
injNHLdata18 = injNHLdata[injNHLdata['Season']== '2018/19']
injNHLdata18 = injNHLdata18.drop(['Games Missed.1'], axis=1)
injNHLdata18.head()
```

```python
#  rename cols/ remove comma in names

injNHLdata18 = injNHLdata18.rename(index=str, columns={"Cap Hit": "CAP HIT", "Chip": "CHIP", "Games Missed": "GAMES MISSED", "Injury Type": "INJURY TYPE", "Player": "PLAYER", "Position": "POSITION", "Season": "SEASON", "Team":"TEAM"})


injNHLdata18['PLAYER'] = injNHLdata18['PLAYER'].str.replace(',', '')
print(injNHLdata18.shape)
injNHLdata18.head()
```

```python
# drop duplicated rows if any
injNHLdata18 = injNHLdata18.drop_duplicates(subset=None, keep='first', inplace=False)
# print(injNHLdata18.shape)
injNHLdata18.head()
```

```python
# remove rows with non injury reasons
injNHLdata18 = injNHLdata18[injNHLdata18['INJURY TYPE'] != 'Illness']
injNHLdata18 = injNHLdata18[injNHLdata18['INJURY TYPE'] != 'Flu']
injNHLdata18 = injNHLdata18[injNHLdata18['INJURY TYPE'] != 'Undisclosed']
injNHLdata18.head()
```

```python
# count injuries per player
inj_grpby18 = pd.DataFrame(injNHLdata18.groupby('PLAYER')['INJURY TYPE'].count()).reset_index()
inj_grpby18 = inj_grpby18.rename(index=str, columns={"INJURY TYPE": "INJURY COUNT"})

inj_grpby18.head()
```

```python
# merge injury count df with main df
injCount18=pd.merge(inj_grpby18, injNHLdata18, left_on ='PLAYER',right_on='PLAYER',how="left",sort=True)
injCount18.head()
```

```python
# write to csv

injCount18.to_csv('injCount18.csv', index=True, header=True)
injNHLdata18.to_csv('injNHLdata18.csv', index=True, header=True)
inj_grpby18.to_csv('injCntNHL18.csv', index=True, header=True)

```

```python
import pandas as pd
import urllib.request

from bs4 import BeautifulSoup
```

```python
# scrape players body stats 2018

source = urllib.request.urlopen('https://www.foxsports.com/nhl/players?teamId=0&season=2018&position=0&country=0&grouping=0&weightclass=0&playerName=').read()
soup = BeautifulSoup(source, "html.parser")
table = soup.find_all('table', attrs={'class':'wisbb_standardTable'})
```

```python
c=0
for tb in table:
#     print(tb)
    c=c+1
print(c)
```

```python
first_table=table[0]

table_rows = first_table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
BodstatNHLdf_18 = pd.DataFrame(res)

# print(df_1819.count)
```

```python
BodstatNHLdf_18 = pd.DataFrame(res, columns=["PLAYER", "TEAM", "POSITION", "HEIGHT", "WEIGHT", "BIRTHDAY"])
BodstatNHLdf_18.head()
```

```python
# new data frame with split value columns 
BodstatNHLdf_18["PLAYER"]= BodstatNHLdf_18["PLAYER"].str.split("t", n = 1, expand = True) 
BodstatNHLdf_18.head()
```

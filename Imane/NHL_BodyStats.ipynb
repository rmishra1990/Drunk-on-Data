```python
import pandas as pd
```

```python
# open csv
file1 = "NHL_Bio.csv"
BodstatNHLdf_18 = pd.read_csv(file1)
BodstatNHLdf_18.head()
```

```python
#drop col"Unnamed"
BodstatNHLdf_18 = BodstatNHLdf_18.drop(['Unnamed: 0'], axis=1)
BodstatNHLdf_18.head()
```

```python
# rename columns
BodstatNHLdf_18 = BodstatNHLdf_18.rename(index=str, columns={"Player": "PLAYER","Team": "TEAM", "Position":"POSITION", "Height": "HEIGHT", "Weight":"WEIGHT", "Birthdate": "BIRTHDATE"})
BodstatNHLdf_18.head()
```

```python
# new data frame with split value columns 
BodstatNHLdf_18["PLAYER1"], BodstatNHLdf_18["PLAYER2"] = BodstatNHLdf_18["PLAYER"].str.split("  ", 1).str 

BodstatNHLdf_18.head()
```

```python
# drop cols PLAYER and PLAYER2
BodstatNHLdf_18 = BodstatNHLdf_18.drop(['PLAYER', 'PLAYER2'], axis=1)

BodstatNHLdf_18.head()
```

```python
# rename PLAYER1 col to PLAYER
BodstatNHLdf_18 = pd.DataFrame(BodstatNHLdf_18.rename(index=str, columns={"PLAYER1": "PLAYER"}))
# BodstatNHLdf_18 = BodstatNHLdf_18.iloc[1:]
BodstatNHLdf_18.head()
```

```python
# add/initialize AGE col in df
BodstatNHLdf_18["AGE"] = ""
BodstatNHLdf_18.head()
```

```python
# dedup any replicated rows
BodstatNHLdf_18 = BodstatNHLdf_18.drop_duplicates(subset='PLAYER', keep='first', inplace=False)
BodstatNHLdf_18.head()
```

```python
#Function to remove '-' from specific columns and returns back DF
def clean_my_data(tmpdf):
    tmpdf=tmpdf.loc[tmpdf.WEIGHT != "-"]
    tmpdf=tmpdf.loc[tmpdf.HEIGHT != "-"]
    tmpdf=tmpdf.loc[tmpdf.POSITION != "-"]
    tmpdf=tmpdf.loc[tmpdf.BIRTHDATE != "-"]
    tmpdf=tmpdf.loc[tmpdf.TEAM != "-"]
    tmpdf=tmpdf.loc[tmpdf.PLAYER != "-"]
    
    return(tmpdf)

BodstatNHLdf_18=clean_my_data(BodstatNHLdf_18)
BodstatNHLdf_18.head()
```

```python
# calculate age
import datetime as dt

now = dt.datetime.now()
for i in range(0, len(BodstatNHLdf_18)):
    diff = now - dt.datetime.strptime(BodstatNHLdf_18['BIRTHDATE'][i],'%m/%d/%Y')
    years = diff.days // 365
    days = diff.days - (years*365)
    BodstatNHLdf_18['AGE'][i] = years

# print(BodstatNHLdf_18['AGE'].value_counts())    
BodstatNHLdf_18.head()
```

```python
BodstatNHLdf_18['PLAYER'] = BodstatNHLdf_18['PLAYER'].str.replace(',', '')
BodstatNHLdf_18.head()
```

```python
#write into csv file
BodstatNHLdf_18.to_csv('NHLBodstat_18.csv', index=True, header=True)
```

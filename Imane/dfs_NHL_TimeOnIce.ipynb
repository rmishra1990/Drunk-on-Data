```python
import pandas as pd
import urllib.request

from bs4 import BeautifulSoup
```

```python
# scrape players Time on Ice 2018

source = urllib.request.urlopen('http://www.espn.com/nhl/statistics/player/_/stat/timeonice/sort/avgTimeOnIce/year/2018').read()
soup = BeautifulSoup(source, "html.parser")
table = soup.find_all('table', attrs={'class':'tablehead'})
```

```python
c=0
for tb in table:
#     print(tb)
    c=c+1
print(c)
```

```python
first_table=table[0]

table_rows = first_table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
timeNHL_df_18 = pd.DataFrame(res)

# print(df_1819.count)
```

```python
timeNHL_df_18 = pd.DataFrame(res, columns=["RK","PLAYER", "TEAM", "GP", "G", "A", "PTS", "+/-", "TOI/G", "SHFT", "SHFT/G", "PROD"])
timeNHL_df_18.head()
```

```python
timeNHL_df_18 = timeNHL_df_18.iloc[1:]
timeNHL_df_18
```

```python
timeNHL_df_18 = timeNHL_df_18.drop_duplicates(subset='PLAYER', keep='first', inplace=False)
timeNHL_df_18 = timeNHL_df_18.drop('RK', axis=1)

timeNHL_df_18
```

```python
timeNHL_df_18 = timeNHL_df_18.drop(timeNHL_df_18.index[10])

timeNHL_df_18[['PLAYER','POS']] = timeNHL_df_18['PLAYER'].str.split(',',expand=True)

timeNHL_df_18
```

```python
# flip first last name to last first
timeNHL_df_18["PLAYER"] = [" ".join(n.split(" ")[::-1]) for n in timeNHL_df_18["PLAYER"]]
timeNHL_df_18
```

```python
import pandas as pd
import urllib.request

from bs4 import BeautifulSoup
```

```python
# scrape players Time on Ice 2018

source = urllib.request.urlopen('http://www.espn.com/nhl/statistics/player/_/stat/timeonice/sort/avgTimeOnIce/year/2018').read()
soup = BeautifulSoup(source, "html.parser")
table = soup.find_all('table', attrs={'class':'tablehead'})
```

```python
c=0
for tb in table:
#     print(tb)
    c=c+1
print(c)
```

```python
first_table=table[0]

table_rows = first_table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
timeNHL_df_18 = pd.DataFrame(res)

# print(df_1819.count)
```

```python
timeNHL_df_18 = pd.DataFrame(res, columns=["RK","PLAYER", "TEAM", "GP", "G", "A", "PTS", "+/-", "TOI/G", "SHFT", "SHFT/G", "PROD"])
timeNHL_df_18.head()
```
